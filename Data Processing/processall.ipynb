## Average stars

```
scottsdale = pd.DataFrame()
phoenix = pd.DataFrame()
mesa = pd.DataFrame()
tempe = pd.DataFrame()
maricopa = pd.DataFrame()

scottsdale = final_sentiment[final_sentiment['processed_city']=='Phoenix'].copy().reset_index().drop(['index'],axis=1)
phoenix = final_sentiment[final_sentiment['processed_city']=='Scottsdale'].copy().reset_index().drop(['index'],axis=1)
mesa = final_sentiment[final_sentiment['processed_city']=='Mesa'].copy().reset_index().drop(['index'],axis=1)
tempe = final_sentiment[final_sentiment['processed_city']=='Tempe'].copy().reset_index().drop(['index'],axis=1)
maricopa = final_sentiment[final_sentiment['processed_city']=='Maricopa County'].copy().reset_index().drop(['index'],axis=1)

citydf = [
   scottsdale,
   phoenix,
   mesa,
   tempe,
   maricopa       
]

def computeScore(arr):
  arr_real = ast.literal_eval(arr)
  return sum(arr_real)/len(arr_real)

starcitydf = {}

## compute star score
for city in citydf:
  cityname = city['processed_city'][0]
  city2 = city.copy()
  for i in range(len(city2)):
    city2['user_stars'][i] = computeScore(city2['user_stars'][i])
  
  starcitydf[cityname] = city2.copy()

##Cities remove index
starcitydf['Maricopa County'] = starcitydf['Maricopa County'].drop(['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1'],axis=1)
starcitydf['Mesa'] = starcitydf['Mesa'].drop(['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1'],axis=1)
starcitydf['Scottsdale'] = starcitydf['Scottsdale'].drop(['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1'],axis=1)
starcitydf['Tempe'] = starcitydf['Tempe'].drop(['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1'],axis=1)
starcitydf['Phoenix'] = starcitydf['Phoenix'].drop(['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1'],axis=1)


## Save the files
for city in starcitydf.keys():
  newdf = starcitydf[city].drop(['stars', 'review_count', 'is_open', 'categories',
       'processed_city', 'user_fans', 'checkin_dates',
       'reviews2013', 'reviews2014', 'reviews2015', 'rate_mid_2013',
       'rate_start_2014', 'rate_mid_2014', 'rate_start_2015'],axis=1)
  city_name = city.replace(' ','_')
  newdf.to_csv('preparedata/userstars/'+city_name+'_userStars.csv',header=True,index=True)
```
