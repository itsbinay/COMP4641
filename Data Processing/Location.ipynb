```
import geopandas as gpd
import geopy
import time
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
longtitudes = df3['longitude'].copy()
latitudes = df3['latitude'].copy()
#ss
locationList = []
neighbourhoods = []
for i in range(len(latitudes)):
  if i > (3535+1218+1225+1269):
    if(i % 100 == 0 and i != 0):
      print(i," Sleeping...")
      # time.sleep(5)
      print("Woke up...",end='\n\n')
    coordinates = str(latitudes.iloc[i])
    coordinates += ", "
    coordinates += str(longtitudes.iloc[i])
    locator = Nominatim(user_agent="openmapquest")
    location = locator.reverse(coordinates, timeout=100000)
    # location = RateLimiter(locator.reverse(coordinates, timeout=100000), min_delay_seconds=0.1)
    locationList.append(location)
    keylist = location.raw['address'].keys()
    if('city' in keylist):
      neighbourhoods.append(location.raw['address']['city'])
    elif ('county' in keylist):
      neighbourhoods.append(location.raw['address']['county'])
    else: 
      neighbourhoods.append("NA")
  
```
